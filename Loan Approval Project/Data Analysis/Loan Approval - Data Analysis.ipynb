```python
from scipy.stats import ttest_ind

group1 = df1[df1['loan_status'] == 0]['credit_score']
group2 = df1[df1['loan_status'] == 1]['credit_score']

t_stat, p_val = ttest_ind(group1, group2)
print("T-statistic:", t_stat)
print("P-value:", p_val)
```
